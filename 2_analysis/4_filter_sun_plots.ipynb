```python
import pandas as pd
import os
import subprocess
from pathlib import Path
pd.set_option("display.max_columns", None)
```

```python
path_target = '../data/res/res_prod_target.csv'
newdir = '/home/juan/Documents/dev/bio/mirna_mite/data/res/sun_plots_small/'
olddir = '/home/juan/Documents/dev/bio/mirna_mite/data/res/sun_plots/'
```

```python
df = pd.read_csv(path_target)
print(len(df.index))
df.head(2)
```

```python
# filter
files = []
for k, v in df.iterrows():
    experiment = v['Tplot_file_path'].split("/")[-2]
    newname = v['Tplot_file_path'].split("/")[-1]
    Path(f"{newdir}{experiment}").mkdir(parents=True, exist_ok=True)
    cmd = f"cp {v['Tplot_file_path']} {newdir}{experiment}/{newname}"
    os.system(cmd)
    #print(cmd)
    #break
```

```python
# check filtered
print(len(files))
files[:2]
```

```python
# save
for file in files:
    experiment = file.split("/")[-2]
    newname = file.split("/")[-1]
    Path(f"{newdir}{experiment}").mkdir(parents=True, exist_ok=True)
    cmd = f"cp {file} {newdir}{experiment}/{newname}"
    os.system(cmd)
    #print(cmd)
    #break
    
```
