```python
import pandas as pd
```

```python
df_us = pd.read_csv('../data/res/res_prod_target.csv', sep=',')
print(len(df_us))
df_us.head(2)
```

```python
df_sun_mirnas = pd.read_csv('../data/sun_res/mirnas.csv')
print(len(df_sun_mirnas))
df_sun_mirnas.head(2)
```

```python
df_us['mirna_seq'] = df_us['MajorRNA'].str.replace("U","T")
df_us.head(2)
```

```python
df_merge = pd.merge(df_us,df_sun_mirnas,left_on='mirna_seq', right_on="mature sequence (5'-3')")
print(len(df_merge.index))
```

```python
df_sun_targets = pd.read_csv('../data/sun_res/targets.csv')
print(len(df_sun_targets))
df_sun_targets.head(2)
```

```python
df_merge_2 = pd.merge(df_merge,df_sun_targets,left_on='family name', right_on="miRNA family")
print(len(df_merge_2.index))
```

```python
cols = ['MajorRNA', 'mirbase', 'miRNA_Acc.', 'Target_Acc.','family name','annotation_target','reference sequence','Location of cleavage site','category','score','annotation_y']
df_merge_2[cols].to_csv('../data/res/sun_targets_validate.csv', index=None)
```
