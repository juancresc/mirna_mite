```python
import pandas as pd
from Bio import SeqIO
from Bio.SeqRecord import SeqRecord
from Bio.Seq import Seq
import os
```

```python
fasta_seq = SeqIO.parse('../data/allmites.fasta', 'fasta')
buffer_seqs = []
for record in fasta_seq:
    if 'MITE' in record.id or 'DTT' in record.id and not 'MERMITE18A' in record.id:
        buffer_seqs.append(record)
SeqIO.write(buffer_seqs, '../data/allmites.filtered.fasta', "fasta")
```

```python
clus_files = '../data/mites/'
outfile = '../data/res/mites_consensus.fasta'
```

```python
clusters = [f for f in os.listdir(clus_files) if os.path.isfile(os.path.join(clus_files, f))]
```

```python
buffer = []
dismissed = 0
founds = 0
news = 0
mite_count = 0
for cluster in clusters:
    c_dir = clus_files + cluster 
    fasta_seq = SeqIO.parse(c_dir, 'fasta')
    found = False
    count = 0
    dismiss = False
    seqname = ''
    for record in fasta_seq:
        if len(str(record.seq)) > 1100:
            dismiss = True
            seqname = record.id
            break
        if not 'MITE_T' in record.id:
            founds += 1
            found = True
            seqname = record.id
        count += 1
    if not found and not dismiss:
        seqname = 'MITE_' + str(mite_count) + ' ' + record.id
        print(cluster, count, 'new', seqname)
        mite_count += 1
        news += 1
    if found and not dismiss:
        founds += 1
        print(cluster, count, seqname)
    if dismiss:
        dismissed += 1
        print(cluster, count, 'dismiss', seqname)
    else:
        seq = SeqRecord(record.seq, id=seqname, description="_")
        buffer.append(seq)
```

```python
SeqIO.write(buffer, outfile, "fasta")
outfile
```
